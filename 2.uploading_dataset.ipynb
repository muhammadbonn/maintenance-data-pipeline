### Uploading Dataset from Kaggle to AWS S3 Bucket

```python
import kagglehub
import os
import boto3
from pyspark.sql import SparkSession
```

```python
AWS_ACCESS_KEY = "**********"
AWS_SECRET_KEY = "********************"
AWS_REGION = "eu-north-1"
BUCKET_NAME = "maintenance-bonn-2026" 

print("Connecting to AWS S3...")
s3_client = boto3.client(
    's3',
    aws_access_key_id=AWS_ACCESS_KEY,
    aws_secret_access_key=AWS_SECRET_KEY,
    region_name=AWS_REGION
)
print("Connected!")
```

```python
print("Pulling Data directly from Kaggle...")
dataset_path = kagglehub.dataset_download("stephanmatzka/predictive-maintenance-dataset-ai4i-2020")
print(f"Data temporarily cached at: {dataset_path}")

csv_file = None
for file in os.listdir(dataset_path):
    if file.endswith(".csv"):
        csv_file = os.path.join(dataset_path, file)
        break

# Uploading directly to AWS S3
if csv_file:
    s3_destination = "bronze/ai4i2020_raw.csv" 
    print(f"Pushing directly to S3 at: {s3_destination} ...")
    s3_client.upload_file(csv_file, BUCKET_NAME, s3_destination)
    print("Magic Done! Kaggle data is now natively sitting in S3 Bronze Layer.")
else:
    print("Could not find the CSV file!")
```
